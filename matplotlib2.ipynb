```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt



# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")

df = dataset['train'].to_pandas()


# Data Cleanup
df['job_posted_date']= pd.to_datetime(df["job_posted_date"])
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df.job_posted_date.head()
```

```python
date_counts = df.job_posted_date.value_counts()
date_counts
```

```python
date_counts = date_counts.sort_index()
```

```python
plt.plot(date_counts.index,date_counts.values)
```

```python
df['job_posted_date'] = df['job_posted_date'].dt.month
df
```

```python
monthly_counts = df.job_posted_date.value_counts()

monthly_counts = monthly_counts.sort_index()

monthly_counts
```

```python
job_counts = df.job_title.value_counts()
job_counts
```

```python
job_counts = df['Job Title'].value_counts()

plt.bar(job_counts.index, job_counts)
plt.title('Posting by Job Title')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Job Title')
plt.ylabel('Count Of Job')
plt.show()
```
